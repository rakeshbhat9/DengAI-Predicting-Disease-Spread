#### **SelectFromModel**

```
#Estimators - Consolidated from SKlearn cheatsheet based on data at hand.
est = dict()
est['lin'] = LinearRegression()
est['lassocv'] = LassoCV()
est['lasso'] = Lasso()
est['rfr'] = RandomForestRegressor()
est['ridge'] = Ridge()
```

```
# iq data
iq_d = dict()
for key,value in est.items():
  X = iq_data.drop('total_cases',axis = 1)
  y = iq_data['total_cases']
  clf = value.fit(X,y)
  print"-----------------------------------------------------------------------------"

  print "Fitting "+key+" model"

  model = SelectFromModel(clf, prefit=True)
  
  
  all_feats = model.get_support()

  selected_feats = [y for x,y in zip(all_feats,X.columns) if x]
 
  print str(len(selected_feats))+ " Features selected as below"
  print selected_feats

  print ""
  iq_d[key] = selected_feats
  print "Dict updated with features"

  feats_removed = [x for x in X.columns if x not in selected_feats]

  print str(len(feats_removed))+ " Features removed as below"
  print feats_removed
  print "\n"*2
  print "---------------------------------------------------------------------------"
```

```
# sj data
sj_d = dict()
for key,value in est.items():
  X = sj_data.drop('total_cases',axis = 1)
  y = sj_data['total_cases']
  clf = value.fit(X,y)
  print"-----------------------------------------------------------------------------"

  print "Fitting "+key+" model"

  model = SelectFromModel(clf, prefit=True)
  
  
  all_feats = model.get_support()

  selected_feats = [y for x,y in zip(all_feats,X.columns) if x]
 
  print str(len(selected_feats))+ " Features selected as below"
  print selected_feats

  print ""
  sj_d[key] = selected_feats
  print "Dict updated with features"

  feats_removed = [x for x in X.columns if x not in selected_feats]

  print str(len(feats_removed))+ " Features removed as below"
  print feats_removed
  print "\n"*2
  print "---------------------------------------------------------------------------"
```
